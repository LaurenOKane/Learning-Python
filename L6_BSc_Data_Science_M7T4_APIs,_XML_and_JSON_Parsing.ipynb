#**3. Interating with APIs in Python**

You may have noticed in the above code you used the **requests** library and then called a **get** function with the requests library.

**The requests Library in Python**

The requests library is one of the most widely used Python libraries for making HTTP requests to a specified URL. It abstracts the complexities of making requests behind a simple API, allowing users to send HTTP/1.1 requests. With it, you can add content like headers, form data, query parameters, and more to your HTTP requests via simple Python libraries.

You can use it to get HTML content, API data, download files and media, and almost anything that can be done over HTTP/HTTPS.

**There are 4 types of requests we can execute.**

1. **GET**: The GET method is used to retrieve data from a server. It's a read-only operation, meaning it generally has no side effects and doesn't alter the state of the server's resources. In the code above, requests.get() performs a GET request to the weather API to fetch data.

2. **POST**: The POST method is used to submit data to a server for processing. Unlike GET, POST can have side effects and may change the state of the server's resources. For example, you might use POST to upload a file, submit form data, or create a new record in a database.

3. **PUT**: The PUT method is used to update a current resource on the server. Like POST, PUT requests can also change the state of the server's resources. However, PUT requests are idempotent, meaning that making the same request multiple times will have the same effect as making it just once.

4. **DELETE**: The DELETE method is used to remove a resource from the server. Like PUT, DELETE requests are idempotent.

# **4. Understanding Returned API Data (JSON)**

## **4.1 So What is JSON and Why it Matters**

**JSON stands for JavaScript Object Notation**. It's a lightweight data interchange format that's easy for humans to read and write and easy for machines to parse and generate. For example:


```
# A simple JSON object as string
json_string = '{"name": "Alice", "age": 30, "city": "London"}'
```

**Why Does It Matter?**
It's text-based and thus easy to read.
Language independent.
It allows us to serialize and transmit data between servers and web apps.

**JSON and Python: A Native Understanding**

Python provides a **built-in module called json** to work with JSON data.

**JSON vs Python Dictionaries**

JSON objects look a lot like Python dictionaries, but there are some differences, like the need for key-value pairs in JSON to be strings.





````python
# Python dictionary
python_dict = {'name': 'Alice', 'age': 30, 'city': 'London'}
print(python_dict)

# JSON object
json_object = '{"name": "Alice", "age": 30, "city": "London"}'
print(json_object)
````

❓🤔 Notice how the **30** above in Python is formatted as an integer (represented as green in colab), whereas in the JSON object it's red (or a string type).

While Python dictionaries and JSON might look similar at first glance, they differ in purpose, language dependency, supported data types, and other characteristics. Python dictionaries are more flexible but are limited to Python. In contrast, JSON is designed for data interchange and works across multiple programming languages but comes with a more restricted set of data types

---
## **4.2 Serialization & Deserialization**
- **Serialization**: The process of converting a data structure or object state to a JSON formatted string.

- **Deserialization**: The reverse of serialization; converting a JSON formatted string to a Python object.

#### **Serialising JSON in Python**

The json module provides two methods for serialisation:

- `json.dumps()`: Serialises Python object to JSON string.
- `json.dump()`: Serialises Python object to a file-like object.

json.dumps(): A Simple Example


````python
import json

# Python dictionary
person = {'name': 'Alice', 'age': 30, 'city': 'London'}

# Serialization
person_json = json.dumps(person)
print(person_json)
````

The `json.dumps()` method has keyword arguments like indent to make the output more readable.

````python
# Using indent for better readability
print(json.dumps(person, indent=4))
````

#### **Deserialising JSON in Python**
Just like serialisation, the json module offers two methods:

- `json.loads()`: Deserialises JSON strings to Python object.
- `json.load()`: Deserialises file-like object to Python object.

````python
# JSON string
person_json = '{"name": "Alice", "age": 30, "city": "London"}'

# Deserialization
person = json.loads(person_json)
print(person)
````

**We use json.load() for loading from a json file as shown below**

```

    # json.load() is used for deserialising a JSON document from a file or file-like object.

```

## **4.3 Accessing and Indexing JSON Values**

Once we have converted a JSON string into a Python object (usually a dictionary or a list), we can access its elements just like we would with any standard Python data structur

````python
# Given JSON object
json_string = '{"name": "Alice", "age": 30, "city": "London"}'
parsed_json = json.loads(json_string)

# Accessing elements
print(parsed_json["name"])  # Output: Alice
print(parsed_json.get("age"))  # Output: 30
````

#### **Accessing JSON Array Elements**
JSON arrays become Python lists after deserialization. You can index them using zero-based indices.

````python
# Given JSON array
json_array_string = '["apple", "banana", "cherry"]'
parsed_json_array = json.loads(json_array_string)

# Accessing elements
print(parsed_json_array[0])  # Output: apple
print(parsed_json_array[-1])  # Output: cherry
````

**Accessing Nested Elements**

For nested JSON structures, you can use chained square brackets [][] or a combination of [] and .get() for nested elements.

````python
import pprint

# Given nested JSON
nested_json_string = '{"person": {"name": "Alice", "age": 30, "address": {"city": "London", "postcode": "SW1A 1AA"}}}'

# Let's print it out in a nicely formatted manner so
json_object = json.loads(nested_json_string)
json_formatted_str = json.dumps(json_object, indent=2)
print(json_formatted_str)
````

#### **Now let's see how we can access our data inside of the nested JSON**

````python
# Accessing nested elements
print(json_object["person"]["name"])  # Output: Alice
print(json_object["person"]["address"]["city"])  # Output: London

# Alternatively we can use get
print(json_object["person"].get("address").get("postcode"))  # Output: SW1A 1AA
````

### **JSON to Structured Data in Pandas**

## **5.1 Reading and XML in Python**

Python provides a built-in library called **ElementTree** for parsing and creating XML data.

It's imported like:


```
import xml.etree.ElementTree as ET
```



Run the code below to download the XML file we'll be using for this demonstration.

````python
!wget -q --show-progress https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/fitconstructor/employees.xml
````

````python
# Import the ElementTree module from xml.etree
import xml.etree.ElementTree as ET

# Parse the 'employees.xml' file and get the root element of the XML tree
tree = ET.parse('employees.xml')
root = tree.getroot()

# Loop through each 'employee' element in the XML tree rooted at 'company'
for employee in root.findall('employee'):
    # Find the 'name' child element under the 'employee' element and get its text
    name = employee.find('name').text
    # Find the 'role' child element under the 'employee' element and get its text
    role = employee.find('role').text
    # Print the name and role of each employee
    print(f'{name} works as {role}.')
````

## **5.2 Modifying Employee Data in XML**
Imagine you want to update an employee's role or add a new employee altogether.

````python
# Loop through each employee element in the XML tree
for employee in root.findall('employee'):
    # Check if the name matches the one we're looking for
    if employee.find('name').text == 'Olvan':
        # Update the role for the employee named 'John'
        employee.find('role').text = 'Senior Developer'

# Save the updated XML back to a file
tree.write('updated_employees.xml')

for employee in root.findall('employee'):
    # Find the 'name' child element under the 'employee' element and get its text
    name = employee.find('name').text
    # Find the 'role' child element under the 'employee' element and get its text
    role = employee.find('role').text
    # Print the name and role of each employee
    print(f'{name} works as {role}.')
````

We can also add new employees like this:

````python
# Create a new 'employee' element under the root ('company') element
new_employee = ET.SubElement(root, 'employee')

# Create and populate the 'name' sub-element under the new 'employee' element
new_name = ET.SubElement(new_employee, 'name')
new_name.text = 'Emily'

# Create and populate the 'role' sub-element under the new 'employee' element
new_role = ET.SubElement(new_employee, 'role')
new_role.text = 'Data Scientist'

# Create and populate the 'department' sub-element under the new 'employee' element
new_department = ET.SubElement(new_employee, 'department')
new_department.text = 'Data Science'

# Save the updated XML, including the new employee, back to a file
tree.write('updated_employees.xml')

for employee in root.findall('employee'):
    # Find the 'name' child element under the 'employee' element and get its text
    name = employee.find('name').text
    # Find the 'role' child element under the 'employee' element and get its text
    role = employee.find('role').text
    # Print the name and role of each employee
    print(f'{name} works as {role}.')
````

## **5.3 Converting Employee Data to Structured Form**
We can convert this XML data into a more structured format like a Pandas DataFrame for easier analysis.

````python
# Import the required libraries
import pandas as pd
import xml.etree.ElementTree as ET

# Parse the 'employees_with_salary.xml' file and get the root element of the XML tree
tree = ET.parse('employees.xml')
root = tree.getroot()

# Initialise an empty dictionary to hold the employee data
data = {'Name': [], 'Role': [], 'Salary': []}

# Loop through each 'employee' element in the XML tree
for employee in root.findall('employee'):
    # Extract and append the 'name' from the current 'employee' element to the 'Name' list in the dictionary
    data['Name'].append(employee.find('name').text)

    # Extract and append the 'role' from the current 'employee' element to the 'Role' list in the dictionary
    data['Role'].append(employee.find('role').text)

    # Extract and append the 'salary' from the current 'employee' element to the 'Salary' list in the dictionary
    salary_element = employee.find('salary')
    data['Salary'].append(salary_element.text if salary_element is not None else 'Unknown')

# Create a Pandas DataFrame from the populated dictionary
df = pd.DataFrame(data)

# Display the DataFrame to show the structured employee data
df.head()
````
