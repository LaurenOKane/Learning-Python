Creating effective Python notebooks is a crucial skill for data scientists, as it allows them to communicate findings and tell a story in a way that is accessible to both technical and non-technical stakeholders. Here is a brief tutorial on how to create compelling and understandable Python notebooks.

Let's get started!

### **1. Structure Your Notebook**
- **Title and Introduction -** Start with a clear title and an introduction section. The introduction should outline the objectives of the analysis, the dataset(s) being used, and any relevant background information. This sets the stage for your audience.

- **Divide into Sections -** Organise your notebook into sections and subsections, each with descriptive headings. Use markdown cells for these headings to make them stand out. Common sections include Exploratory Analysis, Data Preparation, Model Building, Analysis, and Conclusions.

### **2. Write Clear, Concise Code**
- **Code Comments and Documentation -** Comment your code generously to explain why you're doing something, not just what you're doing. Use docstrings for functions to describe their purpose, parameters, and returns.

- **Refactor for Readability -** Break down complex operations into smaller, logical chunks. If a piece of code is used multiple times, consider writing a function. This not only improves readability but also makes your notebook more modular and easier to update.

### **3. Use Markdown for Narration**
- **Tell a Story -** Use markdown cells to narrate your analysis. After each major step or finding, add a markdown cell to explain what was done, why it was done, and what the findings are. This narrative approach helps guide your audience through your analytical process.

- **Highlight Key Findings -** Use markdown features like bold or italic to emphasise key points or findings. Bulleted or numbered lists can make information easier to digest.

###**4. Visualise Data Effectively**
- **Purposeful Visualisations -** Use visualisations to supplement your findings, not just to display data. Choose the type of plot that best represents the story you are trying to tell, whether it's a comparison, a distribution, a relationship, or a composition.

- **Make it Accessible -** Ensure your visualisations are accessible by adding titles, axis labels, and legends where appropriate. Consider the colour choices for readability and for those with colour vision deficiencies.

###**5. Keep Your Notebook Clean**
- **Remove Unnecessary Code -** Delete or hide code that was used for exploratory purposes but is not part of the final story. This includes importing libraries that were not used, experimental code, and any outputs that are no longer relevant.

**Use Code Folding or Collapsing -** Some notebook interfaces allow you to fold or collapse code cells. This can be useful for making long notebooks more navigable, allowing readers to focus on the analysis narrative and expand code cells if they wish to see the details.

###**6. Reproducibility**
- **Data Source -** Clearly document sources of your data and, if possible, provide direct links or instructions on how to access them. This ensures that your analysis can be replicated or audited by others.

- **Environment Details -** Include information about your Python environment, especially the versions of major libraries used in the analysis. This can be critical for reproducing your results.

```python
import pandas as pd

# Load the dataset
hotel_bookings_df = pd.read_csv('https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/hotel_bookings.csv')

# Explore the first few rows of the dataset
hotel_bookings_df.head()
```

### **1.2 Data Quality Checks**

Let's check for missing values and ensure our dataset is clean before we start our visualisations.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
# check for missing values
hotel_bookings_df.isnull().sum()
```

**Obersvations**

We see most of data has no missing values, except Agent and Companny. Here are the data cleaning steps we'll take:

- Replaced missing values with defaults to indicate:
  - No agency (agent: 0)
  - Private booking (company: 0)
  - Assumed 0 for missing children
  - Marked missing country data as "Unknown"
- Categorised "Undefined" meal options as self-catering (SC)
- Removed entries with no guests (0 adults, 0 children, 0 babies) to maintain data relevance and accuracy

```python
# Define replacements for NaN values to clean the data
# - 'children': Assume 0.0 if missing
# - 'country': Label missing countries as "Unknown"
# - 'agent' and 'company': Assume 0 if missing, indicating no agency/company was used
nan_replacements = {"children": 0.0, "country": "Unknown", "agent": 0, "company": 0}

# Apply the NaN replacements to the dataset
full_data_cln = hotel_bookings_df.fillna(nan_replacements)

# Replace 'Undefined' meal category with 'SC' (Self Catering)
full_data_cln["meal"].replace("Undefined", "SC", inplace=True)

# Identify bookings with no guests (0 adults, 0 children, 0 babies)
zero_guests_indices = full_data_cln[(full_data_cln["adults"] == 0) &
                                    (full_data_cln["children"] == 0) &
                                    (full_data_cln["babies"] == 0)].index

# Drop the bookings with no guests from the dataset
full_data_cln.drop(index=zero_guests_indices, inplace=True)
full_data_cln.head()
```

```python
# How much data is left?
full_data_cln.shape
```

```python
# Separate bookings into Resort Hotel and City Hotel, excluding canceled bookings
rh = full_data_cln[(full_data_cln["hotel"] == "Resort Hotel") & (full_data_cln["is_canceled"] == 0)]
ch = full_data_cln[(full_data_cln["hotel"] == "City Hotel") & (full_data_cln["is_canceled"] == 0)]
```

### **1.4 How much do guests pay for a room per night?**

The room rates at both hotels fluctuate significantly due to variations in room types, meal arrangements, and seasonal influences. Given the absence of explicit currency details and considering Portugal's membership in the European Monetary Union, it is reasonable to infer that all pricing is denominated in Euros (EUR).

**Methodolgy explained below**

In this code, we're calculating the average daily rate per person (excluding babies) for both Resort and City hotels, based on non-canceled bookings. This approach provides insight into the cost efficiency of staying at each hotel type, factoring in only adults and children as paying guests. Further, we normalize the price per person across all bookings and sort them by room type to facilitate a comparative analysis of room pricing strategies between the two hotels. This methodology allows for a more nuanced understanding of pricing dynamics, taking into account the variation in room types and guest compositions

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average daily rate per person (adr_pp) for non-canceled bookings
# This excludes babies from the calculation, considering only adults and children as paying guests
rh["adr_pp"] = rh["adr"] / (rh["adults"] + rh["children"])
ch["adr_pp"] = ch["adr"] / (ch["adults"] + ch["children"])

# Print the average price per night, per person for both hotel types
print("""From all non-canceled bookings, across all room types and meals, the average prices are:
Resort hotel: {:.2f} â‚¬ per night and person.
City hotel: {:.2f} â‚¬ per night and person."""
      .format(rh["adr_pp"].mean(), ch["adr_pp"].mean()))

# Normalize the price per night per person across the full dataset for comparison
full_data_cln["adr_pp"] = full_data_cln["adr"] / (full_data_cln["adults"] + full_data_cln["children"])

# Filter for actual guests (excluding cancellations)
full_data_guests = full_data_cln[full_data_cln["is_canceled"] == 0]

# Create a dataframe with room prices, sorting by room type for easier comparison
room_prices = full_data_guests[["hotel", "reserved_room_type", "adr_pp"]].sort_values("reserved_room_type")

# boxplot:
plt.figure(figsize=(12, 8))
sns.boxplot(x="reserved_room_type",
            y="adr_pp",
            hue="hotel",
            data=room_prices,
            hue_order=["City Hotel", "Resort Hotel"],
            fliersize=0)
plt.title("Price of room types per night and person", fontsize=16)
plt.xlabel("Room type", fontsize=16)
plt.ylabel("Price [EUR]", fontsize=16)
plt.legend(loc="upper right")
plt.ylim(0, 160)
plt.show()
```

The chart above displays the average price per room, categorised by room type, along with the standard deviation. It's important to note that, due to the anonymisation of the data, rooms identified by the same type letter might not correspond to identical room types across different hotels.

### **1.5 How does the price per night vary over the year?**

This code snippet is designed to analyse and visualise how the price per person for a night's stay in hotels varies over different months of the year. It categorises the data by hotel type, displays the prices in chronological order, and illustrates the variability in prices with the standard deviation.

```python
# Extract relevant data
room_prices_monthly = full_data_guests[["hotel", "arrival_date_month", "adr_pp"]].sort_values("arrival_date_month")

# Define the correct order for the months to ensure data is plotted in chronological order
months_ordered = ["January", "February", "March", "April", "May", "June",
                  "July", "August", "September", "October", "November", "December"]
room_prices_monthly["arrival_date_month"] = pd.Categorical(room_prices_monthly["arrival_date_month"], categories=months_ordered, ordered=True)

# Plotting the data: line plot to show room prices over months for different hotel types, with standard deviation
plt.figure(figsize=(12, 8))
sns.lineplot(x="arrival_date_month", y="adr_pp", hue="hotel", data=room_prices_monthly,
             hue_order=["City Hotel", "Resort Hotel"], ci="sd", size="hotel", sizes=(2.5, 2.5))
plt.title("Room Price Per Night and Person Over the Year", fontsize=16)
plt.xlabel("Month", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("Price [EUR]", fontsize=16)
plt.show()
```

### **1.6 Which are the most busy month?**

This code snippet is designed to visualise the average monthly guest counts for both resort and city hotels. It includes a normalisation step to adjust for different data collection periods across months, ensuring that the comparison between months is fair and accurate.

```python
# Purpose: Visualise the average monthly guest counts for resort and city hotels, accounting for data normalization due to different data collection periods.

# Aggregating guest counts per month for each hotel type
resort_guests_monthly_count = rh.groupby("arrival_date_month")["hotel"].count()
city_guests_monthly_count = ch.groupby("arrival_date_month")["hotel"].count()

# Creating DataFrames for visualisation
resort_guest_data = pd.DataFrame({
    "month": resort_guests_monthly_count.index,
    "hotel": "Resort Hotel",
    "guests": resort_guests_monthly_count.values
})

city_guest_data = pd.DataFrame({
    "month": city_guests_monthly_count.index,
    "hotel": "City Hotel",
    "guests": city_guests_monthly_count.values
})

# Combining the data for both hotel types
full_guest_data = pd.concat([resort_guest_data, city_guest_data], ignore_index=True)

# Specifying the correct chronological order for months
months_ordered = ["January", "February", "March", "April", "May", "June",
                  "July", "August", "September", "October", "November", "December"]
full_guest_data["month"] = pd.Categorical(full_guest_data["month"], categories=months_ordered, ordered=True)

# Normalizing guest counts to account for the collection period differences
# July and August data is from 3 years, others are from 2 years
full_guest_data.loc[(full_guest_data["month"] == "July") | (full_guest_data["month"] == "August"), "guests"] /= 3
full_guest_data.loc[~((full_guest_data["month"] == "July") | (full_guest_data["month"] == "August")), "guests"] /= 2

# Visualisation: Line plot of average guest numbers per month by hotel type
plt.figure(figsize=(12, 8))
sns.lineplot(x="month", y="guests", hue="hotel", data=full_guest_data,
             hue_order=["City Hotel", "Resort Hotel"], size="hotel", sizes=(2.5, 2.5))
plt.title("Average Number of Hotel Guests Per Month", fontsize=16)
plt.xlabel("Month", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("Number of Guests", fontsize=16)
plt.show()
```

#### **Guest Trends and Pricing Patterns for City and Resort Hotels:**

**City Hotel**

- **Peak Period**s: Observes higher guest numbers during the spring and autumn months, coinciding with the periods of highest room prices.
- **Summer Trend**: Despite lower prices in July and August, the hotel experiences a decrease in visitor numbers.

**Resort Hotel:**

- **Summer Pricing and Guests**: Experiences a slight decline in guest numbers from June to September, which aligns with the time of highest room prices.
- **Winter Season**: Both hotels report the lowest number of guests during the winter months, indicating a common off-peak period.

This summary highlights the seasonal fluctuations in guest numbers in relation to room pricing at both the City and Resort hotels, revealing distinct patterns of demand and pricing strategies across different times of the year.

### **1.7 How long do people stay at the hotels?**

```python
# Create a DateFrame with the relevant data:
rh["total_nights"] = rh["stays_in_weekend_nights"] + rh["stays_in_week_nights"]
ch["total_nights"] = ch["stays_in_weekend_nights"] + ch["stays_in_week_nights"]

num_nights_res = list(rh["total_nights"].value_counts().index)
num_bookings_res = list(rh["total_nights"].value_counts())
rel_bookings_res = rh["total_nights"].value_counts() / sum(num_bookings_res) * 100 # convert to percent

num_nights_cty = list(ch["total_nights"].value_counts().index)
num_bookings_cty = list(ch["total_nights"].value_counts())
rel_bookings_cty = ch["total_nights"].value_counts() / sum(num_bookings_cty) * 100 # convert to percent

res_nights = pd.DataFrame({"hotel": "Resort hotel",
                           "num_nights": num_nights_res,
                           "rel_num_bookings": rel_bookings_res})

cty_nights = pd.DataFrame({"hotel": "City hotel",
                           "num_nights": num_nights_cty,
                           "rel_num_bookings": rel_bookings_cty})

nights_data = pd.concat([res_nights, cty_nights], ignore_index=True)

#show figure:
plt.figure(figsize=(16, 8))
sns.barplot(x = "num_nights", y = "rel_num_bookings", hue="hotel", data=nights_data,
            hue_order = ["City hotel", "Resort hotel"])
plt.title("Length of stay", fontsize=16)
plt.xlabel("Number of nights", fontsize=16)
plt.ylabel("Guests [%]", fontsize=16)
plt.legend(loc="upper right")
plt.xlim(0,22)
plt.show()
```

```python
avg_nights_res = sum(list((res_nights["num_nights"] * (res_nights["rel_num_bookings"]/100)).values))
avg_nights_cty = sum(list((cty_nights["num_nights"] * (cty_nights["rel_num_bookings"]/100)).values))
print(f"On average, guests of the City hotel stay {avg_nights_cty:.2f} nights, and {cty_nights['num_nights'].max()} at maximum.")
print(f"On average, guests of the Resort hotel stay {avg_nights_res:.2f} nights, and {res_nights['num_nights'].max()} at maximum.")
```

- For the city hotel there is a clear preference for 1-4 nights.
- For the resort hotel, 1-4 nights are also often booked, but 7 nights also stand out as being very popular.

```python
# Create a DateFrame with the relevant data:
res_book_per_month = full_data_cln.loc[(full_data_cln["hotel"] == "Resort Hotel")].groupby("arrival_date_month")["hotel"].count()
res_cancel_per_month = full_data_cln.loc[(full_data_cln["hotel"] == "Resort Hotel")].groupby("arrival_date_month")["is_canceled"].sum()

cty_book_per_month = full_data_cln.loc[(full_data_cln["hotel"] == "City Hotel")].groupby("arrival_date_month")["hotel"].count()
cty_cancel_per_month = full_data_cln.loc[(full_data_cln["hotel"] == "City Hotel")].groupby("arrival_date_month")["is_canceled"].sum()

res_cancel_data = pd.DataFrame({"Hotel": "Resort Hotel",
                                "Month": list(res_book_per_month.index),
                                "Bookings": list(res_book_per_month.values),
                                "Cancelations": list(res_cancel_per_month.values)})
cty_cancel_data = pd.DataFrame({"Hotel": "City Hotel",
                                "Month": list(cty_book_per_month.index),
                                "Bookings": list(cty_book_per_month.values),
                                "Cancelations": list(cty_cancel_per_month.values)})

full_cancel_data = pd.concat([res_cancel_data, cty_cancel_data], ignore_index=True)
full_cancel_data["cancel_percent"] = full_cancel_data["Cancelations"] / full_cancel_data["Bookings"] * 100

# order by month:
ordered_months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
full_cancel_data["Month"] = pd.Categorical(full_cancel_data["Month"], categories=ordered_months, ordered=True)

# show figure:
plt.figure(figsize=(12, 8))
sns.barplot(x = "Month", y = "cancel_percent" , hue="Hotel",
            hue_order = ["City Hotel", "Resort Hotel"], data=full_cancel_data)
plt.title("Cancelations per month", fontsize=16)
plt.xlabel("Month", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("Cancelations [%]", fontsize=16)
plt.legend(loc="upper right")
plt.show()
```

### **2.2 Feature Selection and Data Preparation**

We are identifying relevant numerical and categorical features for our predictive model, excluding certain features to avoid data leakage and ensure the model's generalizability. The dataset is then split into input features (X) and the target variable (y).

```python
# Import necessary libraries for data preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Feature Selection:
# Identifying numerical and categorical features for the model. Certain features are deliberately
# excluded to prevent data leakage and to ensure the model's general applicability across different datasets.
# Features like 'arrival_date_year', 'assigned_room_type', 'booking_changes', 'reservation_status',
# 'country', and 'days_in_waiting_list' are excluded for these reasons.
num_features = [
    "lead_time", "arrival_date_week_number", "arrival_date_day_of_month",
    "stays_in_weekend_nights", "stays_in_week_nights", "adults", "children",
    "babies", "is_repeated_guest", "previous_cancellations",
    "previous_bookings_not_canceled", "agent", "company",
    "required_car_parking_spaces", "total_of_special_requests", "adr"
]

cat_features = [
    "hotel", "arrival_date_month", "meal", "market_segment",
    "distribution_channel", "reserved_room_type", "deposit_type", "customer_type"
]

# Data Preparation:
# Combining features and separating the dataset into features (X) and the target variable (y).
features = num_features + cat_features
X = hotel_bookings_df.drop(["is_canceled"], axis=1)[features]  # Input features
y = hotel_bookings_df["is_canceled"]  # Target variable
```

### **2.3 Data Preprocessing**

Implementing preprocessing steps for both numerical and categorical data. For numerical features, missing values are imputed with a constant value. Categorical features undergo imputation for missing values and one-hot encoding to transform them into a machine-learning-friendly format. These steps are integrated using ColumnTransformer for efficient preprocessing.

```python
# Data Preprocessing Steps:
# Numerical Features:
# Using a constant strategy for imputation to handle missing values, which is logically chosen
# as '0' for numerical columns, considering the nature of the data where missing values can logically be zeros.
num_transformer = SimpleImputer(strategy="constant")

# Categorical Features:
# Handling missing values by imputing with a placeholder 'Unknown' and applying one-hot encoding
# to transform categorical variables into a format that can be provided to ML algorithms.
cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="Unknown")),
    ("onehot", OneHotEncoder(handle_unknown='ignore'))
])

# Combining Preprocessing Steps:
# Integrating the preprocessing steps for both numerical and categorical features using ColumnTransformer,
# facilitating the simultaneous application of different transformations.
preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_transformer, num_features),
        ("cat", cat_transformer, cat_features)
    ]
)
```

### **2.4 Evaluating a few models**

Here we systematically evaluate a selection of machine learning modelsâ€”Decision Tree, Random Forest, Logistic Regression, and XGBoostâ€”by performing cross-validation to assess their accuracy in predicting our target variable, encapsulating the entire process within a pipeline that handles both data preprocessing and model fitting

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
import numpy as np

# Define a list of tuples where each tuple contains the model name and its corresponding initialised object.
# Using a consistent random_state ensures reproducibility.
models_to_evaluate = [
    ("DT_model", DecisionTreeClassifier(random_state=42)),
    ("RF_model", RandomForestClassifier(random_state=42, n_jobs=-1)),
    ("LR_model", LogisticRegression(random_state=42, n_jobs=-1)),
    ("XGB_model", XGBClassifier(random_state=42, n_jobs=-1))
]

# Configure cross-validation settings: 4 folds = 75% training data, 25% validation data.
# Shuffle data to randomize the splits for more reliable cross-validation results.
kfolds = KFold(n_splits=4, shuffle=True, random_state=42)

# Loop through each model to preprocess data, fit the model, and evaluate using cross-validation.
for model_name, model in models_to_evaluate:
    # Combine data preprocessing and model fitting in a single pipeline for simplicity.
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])

    # Perform cross-validation and calculate scores for evaluation.
    cv_scores = cross_val_score(pipeline, X, y, cv=kfolds, scoring="accuracy", n_jobs=-1)

    # Calculate and print the minimum, maximum, mean, and standard deviation of the cross-validation scores.
    print(f"{model_name} cross-validation accuracy: {np.mean(cv_scores):.4f} +/- {np.std(cv_scores):.4f} (std) "
          f"min: {np.min(cv_scores):.4f}, max: {np.max(cv_scores):.4f}")
```

The RandomForest model outperforms other models in our analysis, indicating its effectiveness for the given dataset. Additionally, we conducted hyperparameter optimisation to refine the model's parameters. However, the improvement in accuracy achieved through this optimisation process is marginal, suggesting that the default or initially chosen hyperparameters were already near-optimal for our scenario. This outcome underscores the importance of selecting the right model and the potential limitations of hyperparameter tuning in enhancing model performance.

### **2. 5 Hyperparameter Optimisation**

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Define the enhanced RandomForestClassifier with specified hyperparameters
rf_model_enh = RandomForestClassifier(
    n_estimators=160,
    max_features=0.4,
    min_samples_split=2,
    n_jobs=-1,
    random_state=0
)

# Specify the number of folds for cross-validation
kfolds = 4  # This should be an integer representing the number of folds

# Create the KFold object with the correct 'n_splits' parameter
split = KFold(n_splits=kfolds, shuffle=True, random_state=42)

# Create a pipeline that includes preprocessing and the model
model_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                             ('model', rf_model_enh)])

# Perform cross-validation and calculate the accuracy scores
cv_results = cross_val_score(model_pipe, X, y, cv=split, scoring="accuracy", n_jobs=-1)

# Output the minimum, maximum, mean, and standard deviation of the cross-validation scores
min_score = round(min(cv_results), 4)
max_score = round(max(cv_results), 4)
mean_score = round(np.mean(cv_results), 4)
std_dev = round(np.std(cv_results), 4)
print(f"Enhanced RF model cross-validation accuracy score: {mean_score} +/- {std_dev} (std) min: {min_score}, max: {max_score}")
```
